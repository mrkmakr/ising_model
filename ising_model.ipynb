```python
#ref : http://enakai00.hatenablog.com/entry/20150106/1420538321
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def check_and_flip(field,x,T):
    y=x[0]
    x=x[1]
    
    yp=y+1
    if yp==100:
        yp=0
    xp=x+1
    if xp==100:
        xp=0
    temp = field[yp,x] + field[y-1,x] + field[y,xp] + field[y,x-1]
    p=1.0/(1+np.exp(-temp/T))
    #print(p)
    if np.random.rand() < p:
        return 1
    return -1
```

```python
n=100000
T_list=np.exp(np.linspace(0.001,0.5,n))
field_hist=[]
field=np.random.rand(100,100)
for k in range(n-10):
    T=T_list[-k]
    xr=[np.random.randint(100),np.random.randint(100)]
    field[xr[0],xr[1]]=check_and_flip(field,xr,T)
    if np.mod(k,n/10)==0:
        field_hist.append(field.copy())
```

```python
plt.figure(figsize=[10,5])
for k,f in enumerate(field_hist):
    plt.subplot(2,5,k+1)
    plt.imshow(f)
    plt.gray()
    plt.axis('off')
plt.savefig('ising.png')
```
